```python
rows = state_string.split('\n')
rows = [r.split('\t') for r in rows]
state_dict = {}
for row in rows:
    state_dict[row[0]] = row[1]
```

```python
state_dict
```

```python
def state_abbr(x):
    if pd.isna(x):
        return 'unknown'
    elif len(x) <= 2:
        return x
    else:
        return state_dict[x]


data['state_abbr'] = data['state'].apply(state_abbr)
```
